- Ejericio 2: Crea una función que tome la raíz cuadrada del valor abosulto de la parte entera de x, donde la raíz está multiplicada por la función signo de x. Además, evalua si el resultado es menor, igual o mayor que cero.

*Nota: El número debe ser digitado por el usuario.*

```python
x = float(input("Introduzca un número real: "))
y = np.sign(x)*np.sqrt(np.fabs(x))
print(y)
```

$$
\begin{pmatrix}

\end{pmatrix}
$$

y otra matriz del mismo espacio matricial y tamaño donde las entradas sean obtenidas del usuario. Luego, multiplica las matrices.

```python
m1 = np.array([[3,8],[8,3]])
print("A continuación se te pedirán las entradas de para una matriz 2x2")
m2 = []
for i in range(2):
    auxi=[]
    for j in range(2):
        auxi.append(float(input("Introduce la entrada: (% 1d,% 1d): "%(i+1,j+1))))
    m2.append(auxi)
print(m1)
print("Otra alternativa para imprimir mejor la matriz es la siguiente: ")
for k in range(2): 
    for l in range(2): 
        print("%1d " %m2[k][l], end = " ") 
    print() 
```

```python
m3=np.array([[0,0],[0,0]])
print("La diferencia de las matrices es: ")
for i in range(2):
    for j in range(2):
        m3[i][j]=m1[i][j] - m2[i][j]
print(m3)
```

*Nota: Para que no se junten los gráficos entre sí, se tiene que ajustar el tamño del gráfico entero.*
